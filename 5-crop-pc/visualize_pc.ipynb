```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import proj3d


data = np.load('2023-01-30_11.00.00_png_ultra.svo_pc.npy')

# print(data)
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]


fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
plt.show()


# coordinates
xs = data[:, 0]
ys = data[:, 1]
zs = data[:, 2]
# attribute
# t_low = data["t_low"]

fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(projection='3d')
img = ax.scatter(xs, ys, zs, cmap=plt.hot())
fig.colorbar(img)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()


fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot()
img = ax.scatter(xs, ys, cmap=plt.hot())
fig.colorbar(img)

ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```
